```python
import pyvisa
import time
```

```python
def listar_dispositivos():
    rm = pyvisa.ResourceManager()
    recursos = rm.list_resources()
    print("Dispositivos conectados:")
    for r in recursos:
        print(f"  - {r}")
    return recursos

def leer_multimetro(resource_name):
    try:
        rm = pyvisa.ResourceManager()
        instrumento = rm.open_resource(resource_name)
        instrumento.timeout = 3000  # 3 segundos
        instrumento.write("*IDN?")
        print(f"Conectado a: {instrumento.read()}")
        
        # Medición
        instrumento.write("READ?")
        lectura = instrumento.read()
        return float(lectura)
    except Exception as e:
        print(f"Error al leer de {resource_name}: {e}")
        return None


```

```python
dispositivos = listar_dispositivos()
```

Multímetro Keithley 2110 5 ½

```python
keithley_id = next((d for d in dispositivos if "2110" in d), None)
if keithley_id:
    lectura_k = leer_multimetro(keithley_id)
    print(f"Lectura Keithley 2110: {lectura_k}")
else:
    print("No se detectó el Keithley 2110")
```
